# DATA VISUALIZATION

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
#Reading the dataset
df = pd.read_excel(r'.\clenned_data.xlsx')
df.head()
```

```python
df.info()
```

```python
# Dimension of the dataset
print(np.shape(df))
# dataset has total 3792 rows with 18 columns
```

```python
# Let's summarize the data to see the distribution of data
print(df.describe())
```

```python
print("Types of houses:",len(df['Type'].value_counts()))
df['Type'].value_counts()
```

```python
from wordcloud import WordCloud
```

```python
count_freq=df['Type'].value_counts()
wc=WordCloud(width=500,height=500,background_color='white')
wc.generate_from_frequencies(count_freq)
plt.figure(figsize=(10,10))
plt.imshow(wc)
```

```python
# find out proportion of the diffrent types of house in whole dataset
labels = df['Type'].unique()
fig, ax1=plt.subplots(figsize=(10,5))
plt.pie(df['Type'].value_counts(), autopct='%.1f%%')
ax1.legend(labels, loc = "upper right")
plt.axis('equal')
plt.show()

'''
As we can see that condos are present the most among all other types of house in dataset. The second most popular 
type is Detached house which is about 21.4 %, whilst the house types such as Others, investment, single_family,retail and office are in 
a very few numbers.

'''
```

```python
fig,ax=plt.subplots(figsize=(15,10),ncols=1,nrows=2)
plt.xticks(rotation=90)
ax[0].bar(df['Type'], df['Walk Score'], label='Walk Score')
plt.tight_layout()
ax[0].set_title('ALl houses and their average walkscore according to the type of house')
ax[0].set_xlabel('Type')
ax[0].set_ylabel('Walkscore')
ax[0].legend()

ax[1].bar(df['Type'], df['house_price'], label='house price')
plt.tight_layout()
ax[1].set_title('All houses and their average house price according to the type of house')
ax[1].set_xlabel('Type')
ax[1].set_ylabel('house price')
ax[1].legend()

'''The average walk score is almost the same from all the types of property/house except single_family and retail types which has a little
less numbers. 

The average house price for detached house is very high as compared to others, which is followed by condo and town house. The 
avg. house price for other, retail and single-family homes has a very low.
'''
```

```python
fig,ax=plt.subplots(figsize=(10,10))
box = [df['Size (sq ft)']]
plt.boxplot(box)
ax.set_title('This boxplot visualizes (size in sq ft) of houses')
ax.set_xlabel('Size (sq ft)')
```

```python
fig,ax=plt.subplots(figsize=(15,10))
plt.scatter(df['Parking Total'][:200],df['Stories'][:200])
plt.xticks(rotation = 90)
ax.set_title('This scatterplot visualizes no. of floors (y-axis) and no. of parking spot (x-axis)')
ax.set_xlabel('Parking Spot')
ax.set_ylabel('Floors in building')
'''
From the given scatter plot we can conclude that the house with more floors have less parking space availble while the house 
with less floors (max 1 floor ) has more parking space availble
'''
```

```python
fig,ax=plt.subplots(figsize=(15,15),ncols=1,nrows=4)
plt.xticks(rotation=90)
rooms_name = ["Bedrooms", "Bathrooms", "Rooms","Kitchens"]
color=['cyan', 'red', 'green', 'blue']
for i in range(4):    
    ax[i].bar(df['Type'], df[rooms_name[i]], label=rooms_name[i],color=color[i])
    plt.tight_layout()
    ax[i].set_title('All houses and their average no. of '+rooms_name[i]+' according to the type of house',fontsize=15)
    ax[i].set_xlabel('Type')
    ax[i].set_ylabel(rooms_name[i])
    ax[i].legend()
'''
Detached and multiplex house types has the more rooms, Bedrooms, bathrooms and kitchen as compared to other types.
There is no any single room, bedroom and kitchen available for Investment, retail and office type property.
Number of bedrooms are more in townhouse and semidetached homes as compared to condo types.
'''
```
